```python
 # Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset (flowers data set)
iris = sns.load_dataset("iris")
```

```python
# First 5 rows
iris.head()
```

```python
# Dataset summary (data types, null values)
iris.info()
```

```python
# tells total number of rows and columns 
iris.shape
```

```python
# Column names
iris.columns
```

```python
# Relationship between two features
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species"
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
# Show data distribution of features
iris.hist(figsize=(10,6))
plt.show()
```

```python
# Detect outliers and show data distribution
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Dataset Features")
plt.show()
```
